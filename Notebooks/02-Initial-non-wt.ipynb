### Load Data

```python
# Importing intrinsic libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statannot import add_stat_annotation
```

```python
# Loading Pandas DataFrame:
data2 = pd.read_csv("../Datasets/initial-mutant-mutant.csv")

# Lets subset the data into our CLC and LDC results
data2_ldc = data2[data2.condition == "LDC"]
data2_clc = data2[data2.condition == "CLC"]
```

# Lets now look at the competitions of KO mutant vs KO mutant from LDC conditions: 

### Make a box plot
Plot the individual replicate data points and group the timepoints together for each strain.
First start the the data from competitions flasks incubated in 12-12 LDC (~150 uE)

```python
# Set variables you want to plot
plt.figure(figsize=(10,5))

x = "strain"
y = "percent-mutant"
order=["day 0", "day 2", "day 6"]

swarmplot2ldc = sns.swarmplot(data=data2_ldc, x=x, y=y, hue = "day", dodge=True, 
               palette=sns.color_palette(['black', 'black', 'black']), hue_order=order)

boxplot2ldc = sns.boxplot(data=data2_ldc, x=x, y=y, hue="day", hue_order=order,
                       palette=sns.color_palette(['seagreen', 'coral', 'royalblue']))

# control x and y limits
boxplot2ldc.set(ylim=(0, 1))
swarmplot2ldc.set(ylim=(0, 1))

boxplot2ldc.axes.set_title("LDC",fontsize=20)
boxplot2ldc.set_xlabel("Percent of Strain 1",fontsize=20)
boxplot2ldc.set_ylabel("Day",fontsize=20)
boxplot2ldc.tick_params(labelsize=20)

stat_1936_0644 = add_stat_annotation(boxplot2ldc, data=data2_ldc, x=x, y=y, hue="day",
                    boxPairList=[(("_1936-_0644", "day 0"), ("_1936-_0644", "day 2")),
                                 (("_1936-_0644", "day 2"), ("_1936-_0644", "day 6")),
                                 (("_1936-_0644", "day 0"), ("_1936-_0644", "day 6"))],
                    test='t-test_ind', textFormat='star', loc='outside', verbose=2, fontsize='x-large')

stat_2453_0644 = add_stat_annotation(boxplot2ldc, data=data2_ldc, x=x, y=y, hue="day",
                    boxPairList=[(("_2453-_0644", "day 0"), ("_2453-_0644", "day 2")),
                                 (("_2453-_0644", "day 2"), ("_2453-_0644", "day 6")),
                                 (("_2453-_0644", "day 0"), ("_2453-_0644", "day 6"))],
                    test='t-test_ind', textFormat='star', loc='outside', verbose=2, fontsize='x-large')

# Fix the legand
# Get the handles and labels. 
handles, labels = boxplot2ldc.get_legend_handles_labels()
# When creating the legend, only use the first two elements to effectively remove the last two.
l = plt.legend(handles[0:3], labels[0:3], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize='x-large', title_fontsize='40')
```

### Make it a bar plot

```python
# Set variables you want to plot
plt.figure(figsize=(10,5))

x = "strain"
y = "percent-mutant"
order=["day 0", "day 2", "day 6"]

swarmplot2ldc = sns.swarmplot(data=data2_ldc, x=x, y=y, hue = "day", dodge=True, 
               palette=sns.color_palette(['black', 'black', 'black']), hue_order=order)

barplot2ldc = sns.barplot(data=data2_ldc, x=x, y=y, hue="day", hue_order=order,
                       palette=sns.color_palette(['seagreen', 'coral', 'royalblue']))

# control x and y limits
barplot2ldc.set(ylim=(0, 1))
swarmplot2ldc.set(ylim=(0, 1))
barplot2ldc.axes.set_title("LDC",fontsize=20)
barplot2ldc.set_xlabel("Percent of Strain 1",fontsize=20)
barplot2ldc.set_ylabel("Day",fontsize=20)
barplot2ldc.tick_params(labelsize=20)

stat_1936_0644 = add_stat_annotation(barplot2ldc, data=data2_ldc, x=x, y=y, hue="day",
                    boxPairList=[(("_1936-_0644", "day 0"), ("_1936-_0644", "day 2")),
                                 (("_1936-_0644", "day 2"), ("_1936-_0644", "day 6")),
                                 (("_1936-_0644", "day 0"), ("_1936-_0644", "day 6"))],
                    test='t-test_ind', textFormat='star', loc='outside', verbose=2, fontsize='x-large')

stat_2453_0644 = add_stat_annotation(barplot2ldc, data=data2_ldc, x=x, y=y, hue="day",
                    boxPairList=[(("_2453-_0644", "day 0"), ("_2453-_0644", "day 2")),
                                 (("_2453-_0644", "day 2"), ("_2453-_0644", "day 6")),
                                 (("_2453-_0644", "day 0"), ("_2453-_0644", "day 6"))],
                    test='t-test_ind', textFormat='star', loc='outside', verbose=2, fontsize='x-large')



plt.legend(loc='upper left', bbox_to_anchor=(1.03, 1))
# Fix the legand
# Get the handles and labels. 
handles, labels = boxplot2ldc.get_legend_handles_labels()
# When creating the legend, only use the first two elements to effectively remove the last two.
l = plt.legend(handles[0:3], labels[0:3], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., )
```

### From this, we conclude that our control competition of the clockless 7942_1936 vs the cikA KO 7942_0644 indicated that cikA out competed the clockless strain missing kaiABC. 

### The strain 7942_2453 (encoding pilM) has a LDC fitness value of +1.7 and 7942_0644 (encoding cikA) has a score of 1.4. While we did see an average increase in the strain with the higher fitness score, this was not statistically signifigant due to the variation we observed between biological replicates.
